```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from matplotlib import ticker

Discarded_Weights_1 = np.array( [ 2.41929963e-05,
                                  3.18444278e-05,
                                   4.05896278e-05])
Energies_1          = np.array([ -3344.20652825, 
                                 -3344.20611631, -3344.20544037])
Discarded_Weights_2 = np.array( [ 2.40900715e-05, 
 3.15487523e-05,
 4.05698882e-05])
Energies_2  =  np.array([ -3344.2061837185,-3344.2057608151,  -3344.20505675])
Discarded_Weights_3 = np.array( [2.22e-05,2.96e-05, 3.89e-05]
)
Energies_3          = np.array([-3344.2056537733,-3344.2052418132,-3344.2045559470]
)

Discarded_Weights_4 = np.array([1.86e-05,
 2.49641803e-05,
 3.52513432e-05])

# Energies
Energies_4          =  np.array([-3344.2050846152, -3344.2047357452, -3344.2041471306])
hartree_to_cm = 219474.63  # Hartree to cm^-1 conversion factor
# Linear regression and plotting for each dataset
datasets = [
    (Discarded_Weights_1, Energies_1, "$S$=0"),
    (Discarded_Weights_2, Energies_2, "$S$=1"),
    (Discarded_Weights_3, Energies_3, "$S$=2"),
    (Discarded_Weights_4, Energies_4, "$S$=3")
]
reg_s0 = scipy.stats.linregress(Discarded_Weights_1, Energies_1)
min_energy = reg_s0.intercept  # Use extrapolated energy at M=∞ for S=0
print(f"Extrapolated energy for S=0: {min_energy} Hartree")
energies_plot = [
    (Energies_1 - min_energy) * hartree_to_cm,
    (Energies_2 - min_energy) * hartree_to_cm,
    (Energies_3 - min_energy) * hartree_to_cm,
    (Energies_4 - min_energy) * hartree_to_cm
]
print(energies_plot)
# Create the plot
fig, ax = plt.subplots(figsize=(4.5, 4.5))
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.1E}"))
# ax.ticklabel_format(style='plain', axis='y')
for i, (Discarded_Weights, Energies, label) in enumerate(datasets):
    # Perform linear regression
    reg = scipy.stats.linregress(Discarded_Weights, Energies)
    
    # Extrapolated energy with standard errors
    emin, emax = min(Energies), max(Energies)
    de = emax - emin
    x_reg = np.array([0, Discarded_Weights[-1] + Discarded_Weights[0]])
    
    # Plot data points
    ax.plot(Discarded_Weights, energies_plot[i], 's', mfc='white', mec=f"C{i}", color=f"C{i}", markersize=9, label=label)
    
    # Plot linear fit line
    ax.plot(x_reg, ((reg.intercept + reg.slope * x_reg)-min_energy)* hartree_to_cm, '--', linewidth=2, color=f"C{i}")
    ax.text(0.05, 0.75 + i * 0.05,  # Adjust y-position by 0.03 for each line to avoid overlap
            "$E(M=\\infty) = %.6f $" % (reg.intercept) + " $E_h$",
            color=f"C{i}", fontsize=10, ha='left', transform=ax.transAxes)


# Configure plot limits and labels
# plt.xlim((0, max(Discarded_Weights_1[-1], Discarded_Weights_2[-1], Discarded_Weights_3[-1], Discarded_Weights_4[-1]) + Discarded_Weights_1[0]))
plt.xlim((0, 5e-5))
plt.ylim(0,1600)
# plt.ylim((reg.intercept - de * 0.8, emax + de * 0.5))
plt.xlabel("Largest Discarded Weight", fontsize=10)
plt.ylabel("Sweep Energy (cm$^{-1}$)", fontsize=10)
plt.legend()
plt.subplots_adjust(left=0.17, bottom=0.1, right=0.95, top=0.95)

# Save and display the plot
plt.savefig("cr2_pantazis_all_spin_states_fiedler.pdf", dpi=600)
plt.show()
```
